```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

hitters = pd.read_csv("batting.csv")
hitters
```

```python
hitters.fillna(0, inplace=True)
```

```python
hitters.head()
```

```python
hitters['AVG'] = hitters['H']/hitters['AB']
hitters['AVG'] = hitters['AVG'].round(3)
hitters['AVG'].fillna(0, inplace=True)
hitters
```

```python
hitters["OBP"] = (hitters["H"] + hitters["BB"] + hitters["HBP"]) / (hitters["AB"] + hitters["BB"] + hitters["HBP"] + hitters["SF"])
hitters["OBP"] = hitters["OBP"].round(3)
hitters.fillna(0.000, inplace=True)
hitters
```

```python
hitters["SLG"] =  ((hitters["H"] - hitters["2B"] - hitters["3B"] - hitters["HR"]) -  + (2*hitters["2B"]) + (3*hitters["3B"]) + (4*hitters["HR"]) ) / hitters["AB"]
hitters["SLG"] = hitters["SLG"].round(3)
hitters.fillna(0.000, inplace=True)
hitters
```

```python
hitters['OPS'] = hitters['OBP'] + hitters['SLG']
hitters['OPS'] = hitters['OPS'].round(3)
hitters
```

```python
thousand_ops = hitters[(hitters['OPS'] >= 1.000) & (hitters['G'] >= 50)]
thousand_ops[['playerID', 'teamID','yearID', 'AVG', 'OBP', 'SLG', 'OPS']]
```
